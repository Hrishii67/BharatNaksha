```python
df_state_pop= pd.read_csv('Indian Population Statewise.csv')
df_state_pop.head()
```

```python
ind_shp = gpd.read_file(r"India_State_Boundary.shp")
ind_shp.crs
print(ind_shp.columns)
ind_shp.head()
```

```python
state_mapping = {}
for state in ind_shp['Name']:
    state_mapping[state.upper()] = state
```

```python
df_state_pop_copy = df_state_pop.copy()
df_state_pop_copy['Name'] = df_state_pop_copy['Name'].map(lambda x: state_mapping.get(x, x))
df_state_pop_copy.head()

df_state_pop_copy.to_csv('Indian Population.csv')
```
